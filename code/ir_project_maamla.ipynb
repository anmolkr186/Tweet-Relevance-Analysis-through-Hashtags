```
key = res[-1]._json.keys()
for i in key:
  print(i)
```

```
filename = "tweets.csv"
    
with open(filename, 'w') as csvfile:  
    csvwriter = csv.writer(csvfile)   
    csvwriter.writerow(key)
    for i in res:
      csvwriter.writerow(i._json.values())  
with open('tweets.json', 'w', encoding='utf-8') as f:
  for i in res:
    json.dump(i._json, f, indent = 4, ensure_ascii=False)
```
